ONE HOT ENCODING

```python
df.realtionship_status.unique()
```

```python
one_hot_encoding = OneHotEncoder()
one_hot_encoding = one_hot_encoding.fit(df['realtionship_status'].values.reshape(-1, 1))
```

```python
one_hot_encoding.categories_
```

```python
one_hot_labels = one_hot_encoding.transform(df['realtionship_status'].values.reshape(-1,1)).toarray()
```

```python
labels_df = pd.DataFrame()

labels_df['Divorced'] = one_hot_labels[:,0]
labels_df['Married-AF-spouse'] = one_hot_labels[:,1]
labels_df['Married-civ-spouse'] = one_hot_labels[:,2]
labels_df['Married-spouse-absent'] = one_hot_labels[:,3]
labels_df['Never-married'] = one_hot_labels[:,4]
labels_df['Separated'] = one_hot_labels[:,5]
labels_df['Widowed'] = one_hot_labels[:,6]
```

```python
encoded_df = pd.concat([df, labels_df], axis=1)
encoded_df.drop('realtionship_status', axis=1, inplace=True)
```

```python
encoded_df.head()
```

Airtime

```python
one_hot_encoding = OneHotEncoder()
one_hot_encoding = one_hot_encoding.fit(df['airtime'].values.reshape(-1, 1))
```

```python
one_hot_encoding.categories_
```

```python
one_hot_labels = one_hot_encoding.transform(df['airtime'].values.reshape(-1,1)).toarray()
```

```python
labels_df3 = pd.DataFrame()

labels_df3['Daytime'] = one_hot_labels[:,0]
labels_df3['Morning'] = one_hot_labels[:,1]
labels_df3['Primetime'] = one_hot_labels[:,2]
```

```python
encoded_df3 = pd.concat([encoded_df2, labels_df3], axis=1)
encoded_df3.drop('airtime', axis=1, inplace=True)
```

```python
encoded_df3.head()
```

```python
encoded_df3['ratings']=df['ratings']*100
```

```python
le = LabelEncoder()
encoded_df3['netgain']=le.fit_transform(encoded_df3['netgain'])
```

Splitting Dataset

```python
y = encoded_df3['netgain'].values
```

```python
x = encoded_df3.drop(['netgain'],axis=1).values
```

```python
from sklearn.model_selection import KFold
cv = KFold(n_splits=10,random_state=42,shuffle=False)
for train_index,test_index in cv.split(x):
    print("train index: ",train_index)
    print("test index: ",test_index)
```

```python
x_train,x_test,y_train,y_test = x[train_index],x[test_index],y[train_index],y[test_index]
```

XGBOOST

```python
from xgboost import XGBClassifier
```

```python
alg = XGBClassifier(learning_rate=0.1, n_estimators=140, max_depth=5,
                        min_child_weight=3, gamma=0.2, subsample=0.6, colsample_bytree=1.0,
                        objective='binary:logistic', nthread=4, scale_pos_weight=1, seed=27)
```

```python
alg.fit(x_train,y_train)
```

```python
y_pred_xgboost = alg.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_xgboost))
```
